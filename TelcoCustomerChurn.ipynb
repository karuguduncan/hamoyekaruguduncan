```python
#converting the 'TotalCharges' column to numeric values and filling missing values with 0
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
df['TotalCharges'] = df['TotalCharges'].fillna(0)
```

```python
df['TotalCharges'].info()
```

```python
df.info()
```

```python
df['Churn'].value_counts()
```

```python
#Convert the 'Churn' column to binary values, where 'No' is mapped to 0 and 'Yes' is mapped to 1
mapping = {'Yes': 1, 'No': 0}
df['Churn'] = df['Churn'].map(mapping)

print(df['Churn'].value_counts())
```

```python
x=df.drop(columns='Churn')
y=df['Churn']
```

```python
# Split the data into an 80-20 train-test split with a random state of “1”
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
y_train.value_counts()
```

```python
# categorical = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines',
#'InternetService','OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
#'StreamingMovies','Contract', 'PaperlessBilling', 'PaymentMethod']
# numerical = ['tenure', 'MonthlyCharges', 'TotalCharges']
# we have to sacle the data for better result

categorical_df=df.drop(columns=['tenure', 'MonthlyCharges', 'TotalCharges','Churn'])
numerical_df=df.loc[:,['tenure', 'MonthlyCharges', 'TotalCharges']]
```

```python
# The numerical features should be scaled using StandardScaler,
# convert the output back to a dataframe and put back the column names.
from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()
d_scaled=scaler.fit_transform(numerical_df)
data_scaled1=pd.DataFrame(d_scaled, columns=scaler.get_feature_names_out(numerical_df.columns))
data_scaled1.head()

```

```python
# The categorical features are one-hot encoded using OneHotEncoder(set sparse_output to false),
# convert the output back to a dataframe and put back the column names.

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
encoded_data = encoder.fit_transform(categorical_df)
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_df.columns))

encoded_df.head()
```

```python
# Combine scaled numerical and one-hot encoded categorical features into train and test set dataframes (use pd.concat)
scaled_df=pd.concat([data_scaled1,encoded_df],axis=1)
scaled_df.head()
```

```python

X=scaled_df
y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
y_train.value_counts()
```

```python
# There is still an imbalance in the class distribution. For this, we use SMOTE only on the training data to handle this.
import imblearn
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=1)
x_train_balanced, y_balanced = smote.fit_resample(x_train, y_train)
```

```python
# Use scikit learn to train a random forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report


rf_classifier = RandomForestClassifier(n_estimators=100, random_state=1)
rf_classifier.fit(x_train_balanced, y_balanced)
y_pred1 = rf_classifier.predict(x_test)
y_pred1

accuracyRF = accuracy_score(y_test, y_pred1)
reportRF = classification_report(y_test, y_pred1)

print(f'Accuracy: {accuracyRF}')
print('Classification Report:')
print(reportRF)
```

```python
# extra trees classifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import accuracy_score, classification_report

et_classifier = ExtraTreesClassifier(n_estimators=100, random_state=1)
et_classifier.fit(x_train_balanced, y_balanced)
y_pred2 = et_classifier.predict(x_test)
y_pred2
```

```python
from sklearn.metrics import accuracy_score, classification_report

accuracyET = accuracy_score(y_test, y_pred1)
reportET = classification_report(y_test, y_pred1)

print(f'Accuracy: {accuracyET}')
print('Classification Report:')
print(reportET)
```

```python
# use xgboost to train an extreme boosting model
import xgboost as xgb

xgb_classifier = xgb.XGBClassifier(n_estimators=100, random_state=1)
xgb_classifier.fit(x_train_balanced, y_balanced)
y_pred_xgb = xgb_classifier.predict(x_test)

accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
report_xgb = classification_report(y_test, y_pred_xgb)

print(f'XGBoost Accuracy: {accuracy_xgb}')
print('XGBoost Classification Report:')
print(report_xgb)
```

```python
# use lightgbm to train  a light gradient boosting model
import lightgbm as lgb

lgb_classifier = lgb.LGBMClassifier(n_estimators=100, random_state=1)
lgb_classifier.fit(x_train_balanced, y_balanced)
y_pred_lgb = lgb_classifier.predict(x_test)

y_pred_lgb = lgb_classifier.predict(x_test)
accuracy_lgb = accuracy_score(y_test, y_pred_lgb)
report_lgb = classification_report(y_test, y_pred_lgb)

print(f'LightGBM Accuracy: {accuracy_lgb}')
print('LightGBM Classification Report:')
print(report_lgb)
```
